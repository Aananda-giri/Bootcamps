```python
# Write a python program to implement water jug problem.
'''
Water Jug Problem Solution:
State=(x, y) where x represents the volume of water in 4-gallon jug. y represents the volume of water in 3-gallon jug.

Initial State=(0, 0)    # 0 gallon in 4-gallon jug and 0 gallon in 3-gallon jug

Rules:
(x,y) such that x<4 -> (4,y)    # Fill 4-gallon jug until it is full
(x,y) such that y<3 -> (x,3)    # Fill 3-gallon jug until it is full
(x,y) such that x>0 -> (0,y)     # Empty 4-gallon jug
(x,y) such that y>0 -> (x,0)     # Empty 3-gallon jug
(x,y) such that x+y>=4, y>0 -> (4,y-(4-x))    # Pour water from 3-gallon jug to 4-gallon jug until 4-gallon jug is full
(x,y) such that x+y>=3, x>0 -> (x-(3-y),3)    # Pour water from 4-gallon jug to 3-gallon jug until 3-gallon jug is full
(x,y) such that x+y<=4, y>0 -> (x+y,0)        # Pour all the water from 3-gallon jug to 4-gallon jug
(x,y) such that x+y<=3, x>0 -> (0,x+y)        # Pour all the water from 4-gallon jug to 3-gallon jug
'''
import random
print("******************* Water Jug Problem *******************")
x=0
y=0
state =[x,y]
visited_state = [[0,0]]
print("Initial State: ",state)
while True:
    conflict_set = []
    if (state[0]==2 or state[1]==2):
        print("Goal Achieved: \n")
        break

    if (x<4):
        conflict_set.append(1)
    if (y<3):
        conflict_set.append(2)
    if (x>0):
        conflict_set.append(3)
    if (y>0):
        conflict_set.append(4)
    if ( (x+y) >= 4 and x<4 or y>0):
        conflict_set.append(5)
    if ( (x+y) >=3 and y<3 or x>0):
        conflict_set.append(6)
    if ( (x+y) <= 4 and y>0):
        conflict_set.append(7)
    if ( (x+y) <= 3 and x>0):
        conflict_set.append(8)
    print("conflict set = ", conflict_set)
    while True:
        r = random.choice(conflict_set)
        print("random rule fired = ",r)
        if r==1:
            print("Fill 4-gallon jug until it is full")
            state = [4,y]
            if (state in visited_state):
                print("State already visited")
            else:
                x=4
                print(f"New State=({x}, {y})")
                visited_state.append(state)
                break

        elif r==2:
            print("Fill 3-gallon jug until it is full")
            state = [x,3]
            if (state in visited_state):
                print("State already visited")
            else:
                y=3
                print(f"New State=({x}, {y})")
                visited_state.append(state)
                break

        elif r==3:
            print("Empty 4-gallon jug")
            state = [0,y]
            if (state in visited_state):
                print("State already visited")
            else:
                x=0
                print(f"New State=({x}, {y})")
                visited_state.append(state)
                break

        elif r==4:
            print("Empty 3-gallon jug")
            state = [x,0]
            if (state in visited_state):
                print("State already visited")
            else:
                y=0
                print(f"New State=({x}, {y})")
                visited_state.append(state)
                break
        elif r==5:
            print("Pour water from 3-gallon jug to 4-gallon jug until 4-gallon jug is full")
            state = [4,y-(4-x)]
            if (state in visited_state):
                print("State already visited")
            else:
                y=y-(4-x)
                x=4
                print(f"New State=({x}, {y})")
                visited_state.append(state)
                break
        elif r==6:
            print("Pour water from 4-gallon jug to 3-gallon jug until 3-gallon jug is full")
            state = [x-(3-y),3]
            if (state in visited_state):
                print("State already visited")
            else:
                x=x-(3-y)
                y=3
                print(f"New State=({x}, {y})")
                visited_state.append(state)
                break
        elif r==7:
            print("Pour all the water from 3-gallon jug to 4-gallon jug")
            state = [x+y,0]
            if (state in visited_state):
                print("State already visited")
            else:
                x=x+y
                y=0
                print(f"New State=({x}, {y})")
                visited_state.append(state)
                break
        elif r==8:
            print("Pour all the water from 4-gallon jug to 3-gallon jug")
            state = [0,x+y]
            if (state in visited_state):
                print("State already visited")
            else:
                y=x+y
                x=0
                print(f"New State=({x}, {y})")
                visited_state.append(state)
                break
        else:
            print("Invalid rule")
            break

    print("\n Visited States: ",visited_state)


```
