## part-2: Save

* Harder part of RL: Environment, observation, reward model
* still take very long to train
* SB3 makes it easy to make changes in environment

```python
# pip install gymnasium[box2d]

import gymnasium as gym
from stable_baselines3 import A2C, PPO
import os

models_dir = "models/PPO"
logs_dir = "logs"

if not os.path.exists(models_dir):
    os.makedirs(models_dir)

if not os.path.exists(logs_dir):
    os.makedirs(logs_dir)

env = gym.make("LunarLander-v3")
env.reset()

# Initialize the model (with MlpPolicy = Multi Layer Perceptron)
# model = A2C("MlpPolicy", env, verbose=1, tensorboard_log=logs_dir)    # A2C is an Actor-Critic algorithm
model = PPO("MlpPolicy", env, verbose=1, tensorboard_log=logs_dir)    # PPO is an policy gradient algorithm


TIMESTEPS = 10_000

for i in range(1, 30):
    model.learn(total_timesteps=TIMESTEPS, reset_num_timesteps=False, tb_log_name="PPO")
    model.save(f"{models_dir}/{TIMESTEPS * i}")
    print(f"Saved model to {models_dir}/{TIMESTEPS * i}")


'''
!tensorboard --logdirs=logs # to see tensorboard graphs while training.
'''

'''
episodes = 10

for ep in range(episodes):
    obs = env.reset()
    done=False
    while not done:
        env.render()
        action, _ = model.predict(obs)
        # obs, reward, done, info = env.step(env.action_space.sample()) # random action
        obs, reward, done, info = env.step(action)
        print(reward)
        if done:
            break

'''

env.close()


'''
* docs: https://stable-baselines3.readthedocs.io/en/master/

* There is discrete and multi discrete
  * dicrete: one step choice per step
  * multi-step: 8 individual servos (which is multi discrete problem because theere are multiple servos)
'''
```
